```python
import matplotlib.pyplot as plt
import numpy as np

#relative curves
popt_200 = [-0.58853786,  2.36925842]
vel_ver_200 = np.linspace(3,7)
height_ver_200_log = popt_200[0]*np.log(vel_ver_200)+popt_200[1]
height_ver_200 = np.exp(height_ver_200_log)


popt_210 = [-0.57756141,  2.43079976]
vel_ver_210 = np.linspace(3.5,8.83)
height_ver_210_log = popt_210[0]*np.log(vel_ver_210)+popt_210[1]
height_ver_210 = np.exp(height_ver_210_log)

popt_220 = [-0.71905278,  2.67732977]
vel_ver_220 = np.linspace(4,11)
height_ver_220_log = popt_220[0]*np.log(vel_ver_220)+popt_220[1]
height_ver_220 = np.exp(height_ver_220_log)

popt_230 =[-0.59095092,  2.47071631]
vel_ver_230 = np.linspace(4,12)
height_ver_230_log = popt_230[0]*np.log(vel_ver_230)+popt_230[1]
height_ver_230 = np.exp(height_ver_230_log)

popt_240 = [-0.6856249,   2.19719422]
vel_ver_240 = np.linspace(4,12)
height_ver_240_log = popt_240[0]*np.log(vel_ver_240)+popt_240[1]
height_ver_240 = np.exp(height_ver_240_log)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig,ax = plt.subplots(1,1)
fig.dpi=300
ax.plot(vel_ver_200, height_ver_200, label="200 IPM")
ax.plot(vel_ver_210, height_ver_210, label="210 IPM")
ax.plot(vel_ver_220, height_ver_220, label="220 IPM")
ax.plot(vel_ver_230, height_ver_230, label="230 IPM")
ax.plot(vel_ver_240, height_ver_240, label="240 IPM")
ax.set_ylim((0,6))
ax.set_ylabel("Bead Height (mm)")
ax.set_xlabel("Torch Speed (mm/s)")
ax.grid()
plt.legend()
plt.show()
```
