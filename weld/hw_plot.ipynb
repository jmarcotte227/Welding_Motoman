```python
import weld_dh2v, weld_w2v
import matplotlib.pyplot as plt
import numpy as np

vel = [5,15]
height1 = weld_dh2v.v2dh_loglog(vel, 200)
width1 = weld_w2v.v2w_loglog(vel,200)

height2 = weld_dh2v.v2dh_loglog(vel,160)
width2 = weld_w2v.v2w_loglog(vel,160)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig,ax = plt.subplots(1,1)
fig.dpi=300
ax.plot(width2, height2, label="160 IPM")
ax.plot(width1, height1, label="200 IPM")
ax.set_xlabel("Width (mm)")
ax.set_ylabel("Height (mm)")
ax.legend()
plt.show()
```

```python
vel=np.linspace(5,15,1000)
width1 = weld_w2v.v2w_loglog(vel,200)

width2 = weld_w2v.v2w_loglog(vel,160)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig,ax = plt.subplots(1,1)
fig.dpi=300

ax.plot(vel,width2, label="160 IPM")
ax.plot(vel,width1, label="200 IPM")
ax.legend()
ax.set_xlabel("Torch Speed (mm/s)")
ax.set_ylabel("Width (mm)")
plt.show()
```
