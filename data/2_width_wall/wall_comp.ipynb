```python
import numpy as np
import matplotlib.pyplot as plt
material_param={}
```

```python
material_param['ER_4043'] = {
    #ER 4043
    "100ipm": [-0.62015, 1.84913],
    "110ipm": [-0.43334211,  1.23605507],
    "120ipm": [-0.4275129,   1.25877655],
    "130ipm": [-0.42601101,  1.29815246],
    "140ipm": [-0.5068, 1.643],
    "150ipm": [-0.44358649,  1.37366127],
    "160ipm": [-0.4619, 1.647],
    "170ipm": [-0.44802126,  1.40999683],
    "180ipm": [-0.3713, 1.506],
    "190ipm": [-0.5784, 1.999],
    "200ipm": [-0.5776, 2.007],
    "210ipm": [-0.5702, 1.990],
    "220ipm": [-0.5699, 1.985],
    "230ipm": [-0.5374, 1.848],
    "240ipm": [-0.46212367,  1.14990345],
}

def v2dh_loglog(v,mode=140,material='ER_4043'):
    
    mode=int(mode)
    #print(str(mode)+'ipm')
    param = material_param[material][str(mode)+'ipm']
    logdh = param[0]*np.log(v)+param[1]
    
    dh = np.exp(logdh)-0.05
    return dh

def dh2v_loglog(dh,mode=140,material='ER_4043'):
    
    mode=int(mode)
    param = material_param[material][str(mode)+'ipm']

    logdh = np.log(dh)
    logv = (logdh-param[1])/param[0]
    
    v = np.exp(logv)
    return v
```

```python
vels = np.linspace(12,15,100)

plt.plot(vels,v2dh_loglog(vels,200))
plt.plot([12,15],[1.587,1.587],'r')
plt.scatter(13.75,1.587)
```
